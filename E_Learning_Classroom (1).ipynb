```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pywaffle import Waffle
import seaborn as sb

%matplotlib inline
```

The Class was project-based and the evaluation of the skills followed the so-called "21th Century Skills", in a scale from 0 to 10 each Skill:

- Critical Thinking and Problem Solving Skills - named as SK1;
- Creativity and Innovation Skills - named as SK2;
- Constant and Self-Learning Skills - named as SK3;
- Collaboration and Self-Direction Skills - named as SK4;
- Social and Cultural Responsability - named as SK5.


```python
df = pd.read_csv('/Users/laurenthom/Desktop/E-Learning_Classroom/online_classroom_data.csv')
df.head()
```

```python
columns = ['sk1_classroom', 'sk2_classroom', 'sk3_classroom', 'sk4_classroom', 'sk5_classroom' ]
for col in columns:
  df[col] = df[col].str.replace(',', '.').astype(float)
  
df.head()
```

```python
df.describe()
```

```python
df = df.drop(columns=['Approved'])
df.head()
```

```python
cols_to_sum = df.columns[10 : df.shape[1]]

df['score_total'] = df[cols_to_sum].sum(axis=1)

df.head()
```

```python
df.score_total.describe()
```

```python
top_student = df[df['score_total'] == df.score_total.max()]
top_student
```

```python
rows = ['helpful_post','collaborative_post', 'creative_post','confused_post','bad_post']
val = []

for col in rows:
  for v in top_student[col]:
    val.append(v)

sum(val)
```

```python
fig = plt.figure(
    FigureClass= Waffle,
    rows=len(rows),
    values=val,
    colors=["yellow", "blue", "purple", "black", 'brown'],
    icons=['smile','handshake','palette', 'question', 'frown'],
    font_size=20,
    icon_style='solid',
    icon_legend=True,
    legend={
        'labels':['helpful', 'collaborative', 'creative','confusing', 'bad'], 
        'loc': 'upper left', 
        'bbox_to_anchor': (1, 1)
    }
)

plt.title('Student Ratings of Posts for Top Student');


```

```python
low_score_student = df[(df['score_total'] < 35.0) & (df['score_total'] > 30)]
ls_stu = round(low_score_student.mean())
ls_stu.helpful_post
```

```python
low_score_student.shape[0]
```

```python
rows = ['helpful_post','collaborative_post', 'creative_post','confused_post','bad_post']
values = []

for col in rows:
    values.append(ls_stu[col])

sum(values)
```

```python
fig = plt.figure(FigureClass=Waffle, rows=len(rows),values=values,
                 colors=["yellow", "blue", "purple", "black", 'brown'],
                 icons=['smile','handshake','palette', 'question', 'frown'],
                 font_size=20,
                 icon_style='solid',
                 icon_legend=True,
                 legend={
        'labels': ['helpful', 'collaborative', 'creative','confusing', 'bad'], 
        'loc': 'upper left', 
        'bbox_to_anchor': (1, 1)
    }
    
)   

plt.title('Average Student Ratings of Posts for Low Score Students');

```
